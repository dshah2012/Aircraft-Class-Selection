```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```
data=pd.read_csv("train.csv")
data.head(5)
```

```
choice=data['fare_choice']
ori=data['origin']
s=pd.Series(ori)
labels,levels=pd.factorize(s)
```

```
levels
```

```
tempdes=data['destination']
des=[]
for i in range(0,len(tempdes)):
  if tempdes[i]=='MAA':
    des.append(0)
  elif tempdes[i]=='DEL':
    des.append(1)
  elif tempdes[i]=='BOM':
    des.append(2)
  elif tempdes[i]=='BLR':
    des.append(3)
  elif tempdes[i]=='HYD':
    des.append(4)
  elif tempdes[i]=='GOI':
    des.append(5)
  elif tempdes[i]=='CCU':
    des.append(6)
   
```

```
difference=(data['dep_date']-data['booking_date'])
final=data.drop(['id','fare_choice','booking_date','dep_date','origin','destination'],axis=1)
final=final.assign(diff=pd.Series(difference))
final=final.assign(ori=pd.Series(labels))
final=final.assign(de=pd.Series(des))

```

```
final.head(5)
```

```
X_train, X_test,y_train,y_test=train_test_split(final, choice, test_size=0.2, random_state=42)
```
